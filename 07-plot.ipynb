```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0,2 * np.pi,200)
y = np.sin(x)


fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,2,3]
fig.ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('Ientry a')
ax.set_ylabel('entry b')
plt.show
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']=['simHei'] #
plt.rcParams['axes.unicode_minus']=False #

year = [2017,2018,2019,2020]
people = [20,40,60,70]
#
plt.scatter(year,people)
plt.xlabel('')
plt.ylabel('')
plt.title('')

#xyy
plt.fill_betweenx(people,year,color='yellow')

#
plt.show()
```
