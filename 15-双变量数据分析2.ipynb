```python
import pandas as pd
import mytools 
```

```python
# 假设数据在"国家认同数据清理后.sav"文件中  
data = pd.read_spss(R"C:\Users\ceshi\Desktop\SINOOO\国家认同数据清理后.sav")
```

```python
# 创建民族、年级和政治面貌的交互表
cross_tab = pd.crosstab(index=[data['年级'], data['您觉得自己是个典型的中国人吗']], 
                        columns=data['政治面貌'], 
                        margins=True, normalize=True)  # All 行和列的总和

cross_tab
```

```python
from scipy import stats
# 假设我们关注的是第二年级的学生（假设年级用数字“2”表示）
specific_grade = '大二'

# 筛选出特定年级的数据
df_specific_grade = data[data['年级'] == specific_grade]

# 创建会打多少分和政治面貌的交互表
cross_tab_specific_grade = pd.crosstab(df_specific_grade['会打多少分'], df_specific_grade['政治面貌'])

# 进行卡方检验
chi2_grade, p_grade, dof_grade, expected_grade = stats.chi2_contingency(cross_tab_specific_grade)

chi2_grade, p_grade, dof_grade, 
```

```python
# 双变量统计分析 
# 假设我们要分析“政治面貌”和“您觉得自己是个典型的中国人吗”两个变量之间的关系，使用卡方检验  
chi2_result = data.groupby(['政治面貌', '您觉得自己是个典型的中国人吗']).size().reset_index(name='counts').pivot_table(index='政治面貌', columns='您觉得自己是个典型的中国人吗', values='counts').reset_index()  
chi2_test = stats.chi2_contingency(chi2_result)  # 进行卡方检验  
print(f"卡方检验结果：{chi2_test}")
```
